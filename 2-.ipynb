```python
def orcamento_familiar():
    ganhos = float(input("Digite sua renda mensal: "))
    gastos = float(input("Digite suas despesas mensais: "))
    if gastos > ganhos:
        print ("Você está fora do orçamento. Não gaste mais!")
    else:
        print("Você está dentro do orçamento!")
#orcamento_familiar()

def semaforo ():
    cor = input("Digite a cor do semáforo (vermelho, amarelo ou verde): ")
    if cor == "vermelho":
        print("Espere")
    elif cor == "amarelo":
        print("Atenção")
    elif cor == "verde":
        print("Atravesse")
    else:
        print("Farol inoperante")
#semaforo()

def dias_extenso():
    dia = int(input("Digite um número de 0 a 6 para representar um dia da semana: "))
    if dia == 0:
        print("Domingo")
    elif dia == 1:
        print("Segunda-feira")
    elif dia == 2:
        print("Terça-feira")
    elif dia == 3:
        print("Quarta-feira")
    elif dia == 4:
        print("Quinta-feira")
    elif dia == 5:
        print("Sexta-feira")
    elif dia == 6:
        print("Sábado")
    else:
        print("Número inválido. Digite um número entre 0 e 6.")
#dias_extenso()

def sorveteria():
    peso = float(input("Digite o peso do sorvete em gramas: "))
    if peso == 0:
        print("Peso inválido.")
    elif peso < 1000:
        preço = peso * 3.5 / 100
        print ( f"O total é R${preço: .2f}.")
    else:
        preço = peso /100 * 3
        print (f"O total é R${preço: .2f}.")
#sorveteria()

def boletim_aluno ():
  falta = int(input("Digite a quantidade de falta: "))
  n1 = float(input("Digite a nota da prova 1:"))
  n2 = float(input("Digite a nota da prova 2:"))
  n3 = float(input("Digite a nota da prova 3:"))
  media = (n1 + n2 + n3) /3

  if falta < 0 or media < 0 or media > 10:
    print("Dados inválidos")
  elif falta >= 4:
    print (f'''
Média: {media}
Situação: Reprovado por Falta
    ''')
  elif falta < 4 and media > 8 and media <= 10:
    print (f'''
Média: {media}
Situação: Aprovado com sucesso
    ''')
  elif falta < 4 and media < 6:
    print (f'''
Média: {media}
Situação: Recuperação
    ''')
  elif falta < 4 and media == 0:
    print (f'''
Média: {media}
Situação: Desistente
    ''')
  elif falta < 4 and media >= 6 and media <= 8:
    print (f'''
Média: {media}
Situação: Aprovado
    ''')

#boletim_aluno()

def mistura_cor():
    cor1 = input("Digite a primeira cor (vermelho, amarelo ou azul): ")
    cor2 = input("Digite a segunda cor (vermelho, amarelo ou azul): ")

    if (cor1 == "vermelho" and cor2 == "amarelo") or (cor1 == "amarelo" and cor2 == "vermelho"):
        print("A mistura de vermelho e amarelo resulta em laranja.")
    elif (cor1 == "vermelho" and cor2 == "azul") or (cor1 == "azul" and cor2 == "vermelho"):
        print("A mistura de vermelho e azul resulta em roxo.")
    elif (cor1 == "amarelo" and cor2 == "azul") or (cor1 == "azul" and cor2 == "amarelo"):
        print("A mistura de amarelo e azul resulta em verde.")
    else:
        print("Cores inválidas. Digite apenas vermelho, amarelo ou azul.")
#mistura_cor()

def nota_corte():
    nota = float(input("Digite a nota do aluno: "))
    corte = float(input("Digite a nota de corte: "))
    minima = float(input("Digite a nota mínima para aprovação: "))
    if nota < 0 or nota > 10:
        print("Nota inválida. Digite um valor entre 0 e 10.")
    elif nota < minima and nota > corte:
        print("Lista de espera")
    elif nota >= corte:
        print("Aprovado")
    else:
        print("Reprovado")
#nota_corte()

def calculadora():
    n_1 = float(input("Digite o primeiro número: "))
    n_2 = float(input("Digite o segundo número: "))
    operador = input("Digite a operação: ")

    if operador == "+" or operador == "soma":
        resultado = n_1 + n_2
        print(f"O resultado da adição é: {resultado}")
    elif operador == "-" or operador == "subtração":
        resultado = n_1 - n_2
        print(f"O resultado da subtração é: {resultado}")
    elif operador == "*" or operador == "multiplicação":
        resultado = n_1 * n_2
        print(f"O resultado da multiplicação é: {resultado}")
    elif operador == "/" or operador == "divisão":
        if n_2 != 0:
            resultado = n_1 / n_2
            print(f"O resultado da divisão é: {resultado}")
        else:
            print("Erro: Divisão por zero não é permitida.")
    else:
        print("Operação inválida. Digite +, -, *, / ou o nome da operação.")
#calculadora()

def mensalidade_faculdade():
  curso = input("Digite a sigla do curso:")
  if curso == "SI":
    nome = "Sistema de Informação"
    valor = 900
  elif curso == "ADS":
    nome = "Análise e Desenvolvimento de Sistemas"
    valor = 750
  elif curso == "CS":
    nome = "Ciências da Computação"
    valor = 1150
  elif curso == "EC":
    nome = "Engenharia da Computação"
    valor = 1300
  elif curso == "ES":
    nome = "Engenharia de Software"
    valor = 950
  else:
    print("Curso não encontrado")
    return

  isento = input("Você é isento da mensalidade?(Sim ou Não) ")
  if isento == "Sim":
    valor = 0

  desconto = int(input("Qual a porcentagem(%) do desconto: "))
  if desconto <= 0:
    valor = valor
  elif desconto >0:
    valor = valor * desconto / 100

  print (f'''
Curso: {nome}
Valor da mensalidade:
R${valor: .2f}
''')

#mensalidade_faculdade()

def ingresso_cinema():
  inteiras = int(input("Digite a quantidade de ingressos Inteiras: "))
  meias = int(input("Digite a quantidade de ingressos Meias: "))
  dia = input("Digite o dia da semana: ")
  nacional = input("O filme é nacional (Sim ou Não)? ")

  if nacional == "Sim" or nacional == "sim" or nacional == "SIM":
    valor = 5
    print(f'Total à pagar: R${valor * inteiras + valor * meias: .2f}')
  elif dia == "Quarta" or dia == "Quarta-feira" or dia == "quarta-feira" or dia == "quarta":
    valor = 14.5
    print(f'Total à pagar: R${valor * inteiras + valor * meias: .2f}')
  else:
    valor = 28.5
    print(f'Total à pagar: R${valor * inteiras + (valor / 2) * meias: .2f}')

ingresso_cinema()
```
