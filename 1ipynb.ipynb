```python
def faculdade():
    nome = input("Digite seu nome: ")
    curso= input("Digite seu curso: ")
    semestre = input("Digite seu semestre atual: ")
    hobby = input("Digite seu hobby favorito: ")
    print(f'Prazer, eu sou o(a) {nome}!\nAtualmente estou no {semestre}° semestre de {curso} e meu hobby favorito é {hobby}.')
#faculdade()

def churrasco ():
    pessoas = int(input("Digite quantas pessoas irão: "))
    carne = 0.3 * pessoas
    linguica = 0.2 * pessoas
    frango = 0.15 * pessoas
    custo_c = (carne * 50.0)
    custo_l = linguica * 28.0
    custo_f = frango * 22.0
    custo_total = (custo_c + custo_f + custo_l)
    custo_individual = custo_total / pessoas
    print (f'''
    Quantidades:
    Carne:{carne: .1f}Kg - Linguiça:{linguica: .1f}Kg - Frango:{frango: .1f}Kg

    Custo total:
    Carne:R${custo_c: .2f} - Linguiça:R${custo_l: .2f} - Frango:R${custo_f: .2f}

    Custo total do churrasco: R${custo_total: .2f}
    Cada pessoa deve contribuir com: R${custo_individual: .2f}
           ''')
#churrasco()

def media_alunos():
    nome = input("Digite seu nome: ")
    n1 = float(input("Digite a nota da primeira prova: "))
    n2 = float(input("Digite a nota da segunda prova: "))
    n3 = float(input("Digite a nota da terceira prova: "))
    media = (n1 + n2 + n3) / 3
    print (f'O(A) estudante {nome} ficou com a média{media: .1f}.')
#media_alunos()

def conversao ():
    real = int(input("valor em real: "))
    dolar = real / 5.42
    print (f"""
Valor em real R${real: .2f}.
Valor em dólar ${dolar: .2f}.
""")
#conversao()

def compra_online():
    nome = input("Digite seu nome: ")
    valor = float(input("Qual o valor da compra: "))
    desconto = float(input("Digite a porcentagem do desconto: "))
    valor_desconto = valor - (valor * (desconto / 100))
    print (f'''
    Olá {nome}, sua compra de R${valor: .2f} foi confirmada!
    Foi aplicado um desconto de {desconto}%.
    O total final ficou em R${valor_desconto: .2f}.
''')
#compra_online()

def tempo_leitura():
    nome = input("Digite seu nome: ")
    livro = input("Digite o nome do livro: ")
    n_pagina = int(input("Digite quantas páginas tem o livro: "))
    tempo_seg = int(input("Digite em quantos segundos (em média) você lê uma página:"))
    tempo_horas = (tempo_seg * n_pagina) / 3600
    print (f'{nome}, você finalizará a leitura do livro {livro} em aproximadamente {tempo_horas: .2f} horas')
#tempo_leitura()

def acai():
    p = int(input("Quantidade de açaí pequeno:"))
    m = int(input("Quantidade de açaí médio:"))
    g = int(input("Quantidade de açaí grande:"))
    desconto = int(input("Qual a porcentagem do desconto: "))

    valor = (p*13.5 + m*15 + g*17.5) * desconto /100

    print (f'''
    Seu pedido foi registrado.
     - Açaí P: {p}
     - Açaí M: {m}
     - Açaí G: {g}

    Desconto de {desconto}% aplicado.
    Total R${valor: .2f}
           ''')
#acai()

def performance_piloto ():
    pista = int(input("Quantos metros tem a pista: "))
    voltas = int(input("Quantos voltas tem: "))
    segundos = int(input("Quantos segundos para completar a primeira volta: "))

    distancia = (pista * voltas) / 1000
    tempo = (voltas * segundos) / 60

    print (f'''
Análise Preditiva Concluída
--
Distância total a ser percorrida: {distancia} km.
Previsão de conclusão: {tempo} minutos.
''')
#performance_piloto()

def meta ():
    descricao = input("Digite uma breve descrição da sua meta: ")
    meta = float(input("Digite o valor da sua meta: "))
    salario = float(input("Digite o valor do seu salário mensal: "))
    despesas = float(input("Digite o valor das suas despesas mensais: "))
    saldo = salario - despesas
    economia = saldo * 0.3
    valor_disponivel = saldo - economia
    tempo = meta / valor_disponivel
    print (f'''
Meta: {descricao} ({meta: .2f})
Salário: R${salario: .2f} - Despesas: R${despesas: .2f}

Saldo apos despesas: R${saldo: .2f}
Reserva para fixa (30%): R${economia: .2f}
Valor disponível para a meta: R${valor_disponivel: .2f} por mês.
Prazo estimado para alcançar a meta: {tempo: .2f} meses.

''')
meta()
```
